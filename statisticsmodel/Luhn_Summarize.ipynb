# 클래스

```python
from konlpy.tag import Kkma
from nltk.tokenize import sent_tokenize
import numpy as np

class Luhn_Summarize:
    def __init__(self, doc, min = 0.005, max = 0.5):
        self.doc = doc
        self.min = min
        self.max = max

    def preprocessing(self, tokens):
        for i, token in enumerate(tokens):
            if token in ["(", ")", ",", ".", "‧", "…"]:
                del tokens[i]
        return tokens
    

    def keyword(self):
        kkma= Kkma()
        tokens = kkma.morphs(self.doc)
        tokens = self.preprocessing(tokens)
        tokens_uniq = np.unique(tokens)
        freq = {}
        for w in tokens_uniq:
            freq[w] = tokens.count(w)/len(tokens)
        tokens_needed = []
        for w, fre in freq.items():
            if fre > self.min and fre < self.max:
                tokens_needed.append(w)
        return tokens_needed

    def significant(self, num = 3):

        tokens_needed = self.keyword()

        sent = sent_tokenize(self.doc)
        # 제목 부분 처리
        sentence = sent[0].split('\n') + sent[1:]
        
        try:
            while True:
                sentence.remove('')
        except ValueError:
            pass

        significance = {}
        for sent in sentence:
            tok = kkma.morphs(sent)
            tok = self.preprocessing(tok)
            n_needed = []
            for n, t in enumerate(tok):
                if t in tokens_needed:
                    n_needed.append(n)
            n_window = n_needed[-1] - n_needed[0] + 1
            score = np.square(len(n_needed)) / n_window
            significance[score] = sent
    
        return sorted(significance.items(), reverse=True)[:num]
```

```python
ls = Luhn_Summarize(doc)
ls.significant()
```
